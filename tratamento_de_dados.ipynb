###Construindo Gráficos de colunas com dados categóricos 

    ####Gráfico comparando Albuns e Singles:

```python
df_spotify_youtube_ = pd.read_csv("../Projeto_ICD-Analise_de_Dados_do_Spotify_e_Youtube/Spotify_Youtube.csv")
cat_columns = ['Album_type','Licensed','official_video']
df_spotify_youtube_cat = df_spotify_youtube[cat_columns]
counts=df_spotify_youtube_cat['Album_type'].value_counts()
counts = counts.rename_axis('Album_type').rename(index={'compilation': 'Others'})
sns.barplot(x=counts.index,y=counts.values)
plt.xlabel('Album_type')
plt.ylabel('comparação')
plt.title('Album Types')
plt.show()
```

    ####Gráfico de músicas licenciadas:

```python
df_spotify_youtube_ = pd.read_csv("../Projeto_ICD-Analise_de_Dados_do_Spotify_e_Youtube/Spotify_Youtube.csv")
cat_columns = ['Album_type','Licensed','official_video']
df_spotify_youtube_cat = df_spotify_youtube[cat_columns]
counts=df_spotify_youtube_cat['Licensed'].value_counts()
sns.barplot(x=counts.index,y=counts.values)
plt.xlabel('Licensed')
plt.ylabel('comparação')
plt.title('True ou False')
plt.show()
```

    ####Gráfico de músicas com vídeo oficial:

```python
df_spotify_youtube_ = pd.read_csv("../Projeto_ICD-Analise_de_Dados_do_Spotify_e_Youtube/Spotify_Youtube.csv")
cat_columns = ['Album_type','Licensed','official_video']
df_spotify_youtube_cat = df_spotify_youtube[cat_columns]
counts=df_spotify_youtube_cat['official_video'].value_counts()
sns.barplot(x=counts.index,y=counts.values)
plt.xlabel('official_video')
plt.ylabel('comparação')
plt.title('Official Video')
plt.show()
```
